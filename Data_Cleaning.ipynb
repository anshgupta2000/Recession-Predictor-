#Merging the datasets


```python
#Converting all 'Month' and 'Year' Columns to type int from string
inflation_cpi_df['Year']=inflation_cpi_df['Year'].astype(int)
inflation_cpi_df['Month']=inflation_cpi_df['Month'].astype(int)
unrate_df['Year']=unrate_df['Year'].astype(int)
unrate_df['Month']=unrate_df['Month'].astype(int)
yc_df['Year']=yc_df['Year'].astype(int)
yc_df['Month']=yc_df['Month'].astype(int)
cci_df['Year']=cci_df['Year'].astype(int)
cci_df['Month']=cci_df['Month'].astype(int)
fedfunds_df['Year']=fedfunds_df['Year'].astype(int)
fedfunds_df['Month']=fedfunds_df['Month'].astype(int)
buffett_df['Year']=buffett_df['Year'].astype(int)
buffett_df['Month']=buffett_df['Month'].astype(int)
```

```python
data_frames = [inflation_cpi_df,unrate_df,fedfunds_df, cci_df,buffett_df,yc_df]
df = reduce(lambda left,right: pd.merge(left,right,on=['Month','Year'], how='outer'), data_frames)
```

####According to Forbes (https://www.forbes.com/sites/billconerly/2022/05/12/the-yield-curve-predicts-recession-sometimes-but-not-today/?sh=7dd995365443), "We can measure the yield curve by comparing any two maturities, but the most common measure compares the 10-year Treasury bond to the 3-month Treasury bill. "
##Thus I will be removing all other yield curve data except the data that corresponds to the 10-year Treasury bond and the 3-month Treasury bill.

```python
df = df.drop(['Avg 6 Mo','Avg 1 Yr','Avg 2 Yr','Avg 3 Yr','Avg 5 Yr','Avg 7 Yr','Avg 20 Yr','Avg 30 Yr'], axis = 1)
```

```python
#Handling Buffet Indicator, which has values after 3 month gaps. Giving those '3 month gaps' the value of the month prior to that gap.
for i in range(288, len(df)): #because all indices less than 288 are NaN values for the Buffett Indicator. 
  if math.isnan(df['Market_Cap_to_GDP'][i]):
    df['Market_Cap_to_GDP'][i] = df['Market_Cap_to_GDP'][i-1]
```

#Handling NaN Values 

```python
final_df=df.dropna().reset_index(drop=True)
final_df
```

```python
#Creating a column for labels for Recession in the US
#1 indicates recession was there, 0 indicates was not there
#Source: Wikipedia: https://en.wikipedia.org/wiki/List_of_recessions_in_the_United_States 

final_df['Recession_US'] = final_df.apply(lambda x: 1 if ((x['Year'] == 1990)and(x['Month']>=7)) or ((x['Year'] == 1991)and (x['Month']<=3)) or
((x['Year']==2001)and (x['Month']>=3)and (x['Month']<=11)) or
((x['Year']==2007)and (x['Month']==12)) or 
(x['Year']==2008) or 
((x['Year']==2009)and (x['Month']<=6)) or
((x['Year']==2020)and (x['Month']>=2)and(x['Month']<=4))
else 0, axis=1)
```
